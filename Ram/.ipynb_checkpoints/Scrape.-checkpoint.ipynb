```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
def get_synonyms(word_of_interest):
    
    from bs4 import BeautifulSoup
    import requests

    synonyms = []
    
    r = requests.get('http://www.thesaurus.com/browse/' + word_of_interest)
    if (r.status_code == 200):
        try:
            soup = BeautifulSoup(r.content, 'html.parser')
            relevant = soup.find(class_='relevancy-list')
            synonyms = [words.find(class_="text").text for words in relevant.findAll("li")]
        except:
            pass
        
    return synonyms
```

```python
data_dir = '../../data/'
```

```python
train = pd.read_csv(data_dir + 'train.csv', index_col='id')
test = pd.read_csv(data_dir + 'test.csv', index_col='test_id')
```

```python
train = train.dropna()
train.info(null_counts=True)
```

```python
test = test.fillna('', axis=1)
test.info(null_counts=True)
```

```python
train['text'] = train.question1 + ' ' + train.question2
test['text'] = test.question1 + ' ' + test.question2
train.text = train.text.apply(lambda x : list(set(str(x).split())))
test.text = test.text.apply(lambda x : list(set(str(x).split())))
```

```python
flat_train_text = [item for sublist in train.text for item in sublist]
flat_train_text
```

```python
flat_test_text = [item for sublist in test.text for item in sublist]
flat_test_text
```

```python
text = list(set(flat_train_text + flat_test_text))
text
```

```python
len(text)
```

```python
import string
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer

additional_words = 'like '.split()
punct = list(string.punctuation)
nums = list(string.digits)
stop_words = set(stopwords.words('english') + punct + nums + additional_words)
```

```python
text = list(set(text) - stop_words)
len(text)
```

```python
for punctuation in punct:
    text = [word.replace(punctuation, '') for word in text]
    
for num in nums:
    text = [word.replace(num, '') for word in text]
    
len(text)
```

```python
text = list(set(text))
len(text)
```

```python
text = list(set([word for word in text if word and word[0].islower()]))
len(text)
```

```python
text = list(set([word for word in text if 'www' not in word and 'com' not in word]))
len(text)
```

```python
bank = dict()
i = 0

for word in text:
    
    notif_rate = 1000
    total_size = len(text)
    
    syns = get_synonyms(word)
    if syns:
        bank[word] = syns
        
    i += 1
    
    if not i % notif_rate:
        print('Just hit ' + str(i) +  ' words out of ' + str(total_size) + '!')
        print(str(round(i  * 100/ total_size, 2)) + '% of the way there!')
```
